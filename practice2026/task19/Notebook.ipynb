```python
import re
import os
import matplotlib.pyplot as plt

# Путь к файлу в текущей папке task19
file_path = 'result.txt'

if os.path.exists(file_path):
    x = []
    y = []
    
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            match = re.search(r'^\s*(\d+)\s*->\s*команда\s*(\d+)', line)
            if match:
                x.append(int(match.group(1)))
                y.append(int(match.group(2)))

    # Настраиваем график
    plt.figure(figsize=(11, 6))
    plt.plot(x, y, marker='o', color='#31906e', linewidth=2, label='Очередность вызовов')
    plt.grid(True, linestyle='-', alpha=0.3)

    plt.xlim(0.5, len(x) + 0.5)
    plt.ylim(0.7, 5.3)
    plt.xticks(range(1, len(x) + 1))
    plt.yticks([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])

    plt.title('Чередование команд в планировщике', fontsize=12, pad=15)
    plt.xlabel('Порядковый номер вызова Execute()', fontsize=11)
    plt.ylabel('ID команды', fontsize=11)
    plt.legend(loc='lower right', frameon=True, edgecolor='black')

    plt.tight_layout()
    # Сохраняем график прямо в папку task19
    plt.savefig('plot.png', dpi=300)
    plt.show()
    print("График plot.png успешно сгенерирован в папке task19!")
else:
    print("Ошибка: Сначала запустите 'dotnet test' или 'dotnet run' в task19, чтобы сгенерировать result.txt!")
```
